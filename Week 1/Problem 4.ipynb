```python
df = pd.DataFrame({"Review":reviews, "Sentiment":sentiments})
```

```python
print(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=500, stop_words='english')
X = vectorizer.fit_transform(reviews)
```

```python
vectorizer.get_feature_names_out()
```

```python
from sklearn.model_selection import train_test_split
reviews_train_set, reviews_test_set, sentiments_train_set, sentiments_test_set = train_test_split(X, df["Sentiment"], test_size=0.2, train_size=0.8, random_state=12)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(reviews_train_set, sentiments_train_set)
```

```python
from sklearn.metrics import accuracy_score
sentiments_pred = clf.predict(reviews_test_set)
accuracy = accuracy_score(sentiments_test_set, sentiments_pred)
```

```python
print(accuracy)
```

```python
def predict_review_sentiment(model, vectorizer, review):
    pred = model.predict(vectorizer.transform([review]))
    return pred.item()
```

```python
predict_review_sentiment(clf, vectorizer, "A great watch!")
```
