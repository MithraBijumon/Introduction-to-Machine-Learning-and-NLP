```python
import pandas as pd
messages = pd.read_csv(r"C:\Users\mithr\Downloads\spam.csv", encoding = 'latin-1')
messages = messages.drop(labels = ["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis = 1)
messages.columns = ["Label", "Message"]
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

text = "This is an example showing off stop word filtering."
words = word_tokenize(text)
print(words)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
nltk.download('punkt')
def preprocess(text):
    text = text.lower()
    words = word_tokenize(text)
    filtered = [word for word in words if word.lower() not in stop_words]
    return filtered
```

```python
print(stop_words)
```

```python
messages["Message"] = messages["Message"].apply(preprocess)
```

```python
print(messages.head(n=10))
```

```python
import gensim
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

```python
import gensim.models

sentences = messages["Message"]
model = gensim.models.Word2Vec(sentences=sentences)
```

```python
vec_length = model.wv.vector_size
```

```python
import numpy as np
vectors=[]
for message in messages["Message"]:
    vector = np.zeros(vec_length)
    for word in message:
        try:
            vector+=model.wv[word]
        except KeyError:
            pass
    vector = vector/len(vector)
    vectors.append(vector)
```

```python
print(vectors[0])
```

```python
from sklearn.model_selection import train_test_split
messages_train, messages_test, label_train, label_test = train_test_split(vectors, messages["Label"], train_size=0.8, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
clf= LogisticRegression()
clf.fit(messages_train, label_train)
```

```python
from sklearn.metrics import accuracy_score
label_pred = clf.predict(messages_test)
print("Accuracy:", accuracy_score(label_test, label_pred))
```

```python
def predict_messages_class(model, w2v_model, message):
    vec_length = w2v_model.wv.vector_size
    vector = np.zeros(vec_length)
    for word in message:
        try:
            vector+=w2v_model.wv[word]
        except KeyError:
            pass
    vector = vector/len(vector)
    vector = vector.reshape(1, -1)
    pred = model.predict(vector)
    return pred.item()
```

```python
predict_messages_class(clf, model, "Your package from FedEx is awaiting delivery. Please update your address here")
```
